```python
import requests
from bs4 import BeautifulSoup
import pandas
import urllib3
import random

l=[]
base_url="http://www.masterresellrights.com/products/rights/mrr/"
for page in range(1,438):
    
#     import requests
#     from bs4 import BeautifulSoup
#     import urllib3
#     import random
    user_agent_list = [
        'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_15_5) AppleWebKit/605.1.15 (KHTML, like Gecko) Version/13.1.1 Safari/605.1.15',
        'Mozilla/5.0 (Windows NT 10.0; Win64; x64; rv:77.0) Gecko/20100101 Firefox/77.0',
        'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_15_5) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/83.0.4103.97 Safari/537.36',
        'Mozilla/5.0 (Macintosh; Intel Mac OS X 10.15; rv:77.0) Gecko/20100101 Firefox/77.0',
        'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/83.0.4103.97 Safari/537.36',
    ]
    user_agent = random.choice(user_agent_list)
    headers = {'User-Agent': user_agent}
    
    #check if the fake browsers and urls are working fine
    print(base_url+str(page))
    #i=int(page)
    #print("Request #%d\nUser-Agent Sent:%s\n\nHeaders Recevied by HTTPBin:"%(i,user_agent))
    site=requests.get(base_url+str(page),headers=headers)
    print(site.status_code)
    
    
    soup=BeautifulSoup(site.content, "html.parser")
    allTables= soup.find_all("table")
    urlInitials=("http://www.masterresellrights.com")
    
    for i in range (0,10):
        d={}
        
        allTr=allTables[i].find_all("tr")
        #title
        try: d["Product Title"]=allTr[0].a.text
        except: d["Product Title"]=None
        #overview
        getOverview= allTr[2].find_all('p')
        try: d["Overview"]=getOverview[3].text
        except: d["Overview"]=None
            
        #overview 2nd
        try: d["Overview2"]=getOverview[4].text
        except: d["Overview2"]=None
        #productUrl
        productUrl= allTr[0].a.get('href')
        try: d["Product Url"]=urlInitials+productUrl
        except: d["Product Url"]=None
        #imgUrl
        imgUrl= allTr[2].img.get('src')
        try: d["Img Url"]=urlInitials+imgUrl
        except:  d["Img Url"]=None
        l.append(d)
```
